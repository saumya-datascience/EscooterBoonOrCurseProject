```python
import os
import pandas as pd
from county import counties, states, duplicates, uniques
```

```python
count = 0
add= []

for filename in os.listdir('combine'):
    try:
        if filename.endswith(".csv"): 
            df = pd.read_csv(f"combine/{filename}")
            df = df.loc[df["YEAR"]==12, :]
            add.append(df)
    except FileNotFoundError:
        count+=1
count
```

```python
df = pd.concat(add)
```

```python
df.to_csv('combined_states_agesex.csv')
```
