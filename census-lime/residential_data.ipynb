```python
import pandas as pd
from county import counties, states, duplicates, uniques
```

```python
file = 'cc-est2019-all.csv'
df = pd.read_csv(file)

df.head(15)
```

```python
df = df.loc[df["YEAR"]==12, :]
```

```python
df.head()
```

```python
new_df = df.drop_duplicates(subset=['STNAME', 'CTYNAME']).reset_index()
new_df
```

```python
jeff = new_df.loc[new_df["CTYNAME"]=='Jefferson County', :]
jeff
```

```python
duplicates
```

```python
add = []

count=0

for n in duplicates:
    temp_df = new_df.loc[new_df["CTYNAME"]==n[0], :]
    temp_df = temp_df.loc[temp_df["STNAME"]==n[1], :]
    add.append(temp_df)
    
```

```python
for n in uniques:
    temp_df = new_df.loc[new_df["CTYNAME"]==n, :]
    add.append(temp_df)
clean_df = pd.concat(add)
clean_df
```
