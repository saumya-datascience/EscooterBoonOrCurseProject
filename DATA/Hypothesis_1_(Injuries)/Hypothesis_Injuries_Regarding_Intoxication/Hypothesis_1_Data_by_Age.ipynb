```python
#hospitalization data
data_df = pd.read_csv("NEISS.TXT", delimiter = "	")
data_df.head()
```

```python
age = data_df['Age']
print(max(age))
print(min(age))

df = pd.DataFrame(age)
```

```python
# Create the bins in which Data will be held
  
bins = [2, 20, 30, 60, 80, 94]

# Create the names for the five bins
group_names = ["2 -20", "21 to 30", "31 to 60", "61-80", "80-94"]
```

```python
df["Age Group"] = pd.cut(df["Age"], bins, labels=group_names, include_lowest=True)
df
```

```python
new = df.groupby('Age Group').count()
new
newest = new.reset_index()
newest
```

```python
#Display of injuries caused by alcohol
age_group = newest['Age Group']
age_count = newest['Age']

plt.pie(age_count, labels=age_group, autopct="%1.1f%%")
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
plt.title("Injuries by Age")

plt.savefig('Injuries by Age.png')

plt.show()
```
